### 1. Create catalog & schemas

```python
%sql
create catalog if not exists ecommerce;
use catalog ecommerce;
create schema if not exists bronze;
create schema if not exists silver;
create schema if not exists gold;
```

### 2. Register Delta tables

```python
%sql
create table if not exists bronze.events
using delta
as select * from delta.`/Volumes/workspace/ecommerce/delta/bronze/events`;

create table if not exists silver.events
using delta
as select * from delta.`/Volumes/workspace/ecommerce/delta/silver/events`;

create table if not exists gold.smartphone_performance
using delta
as select * from delta.`/Volumes/workspace/ecommerce/delta/gold/smartphone_performance`
```
